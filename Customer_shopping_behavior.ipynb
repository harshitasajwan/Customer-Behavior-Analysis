```python
#We will find out the median output within each category and will replace thtw ith the missing value , that way if a clothing item is missing its review rating it gets replaced by the median of the review ratig of that category only.
```

```python
df['Review Rating']
```

```python
df['Review Rating']=df.groupby('Category')['Review Rating'].transform(lambda x: x.fillna(x.median()))
```

```python
df['Review Rating'].isnull().sum()
```

```python
df.columns=df.columns.str.lower()
df.columns=df.columns.str.replace(' ','_')
```

```python
df.columns
```

```python
df=df.rename(columns={'purchase_amount_(usd)':'purchase_amount'})
```

```python
df.columns
```

```python
#create age group and divide entities by groups
labels=['young adult','adult','middle aged','senior']
df['age_groups']= pd.qcut(df['age'],q=4,labels=labels)
```

```python
df[['age', 'age_groups']].head(15)
```

```python
#create column purchase_frequency_days
frequency_mapping={
    'Fortnightly':14,
    'Weekly':7,
    'Monthly':30,
    'Quarterly':90,
    'Annually':365,
    'Every 3 Months':90,
    'Bi-Weekly':14
}
df['purchase_frequency_days']=df['frequency_of_purchases'].map(frequency_mapping)
```

```python
df[['purchase_frequency_days','frequency_of_purchases']].head(15)
```
